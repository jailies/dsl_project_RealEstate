# preprocessing

```python
def preprocess_news_data(path_raw_data):
    '''
    path_raw_data : '../../data/news/raw/' excel 파일들이 들어 있는 위치, 다른 파일이 있으면 안됨
    '''
    # 경로에 있는 파일 찾기
    file_list = os.listdir(path_raw_data)
    data = pd.DataFrame(columns=['뉴스 식별자', '일자', '언론사', '기고자', '제목', '통합 분류1', '통합 분류2', '통합 분류3', '사건/사고 분류1',
                                   '사건/사고 분류2', '사건/사고 분류3', '인물', '위치', '기관', '키워드', '특성추출(가중치순 상위 50개)',
                                   '본문', 'URL', '분석제외 여부']).set_index('뉴스 식별자')
    # 파일 불로오기, 불필요 열 제거
    for file_name in file_list:
        raw_data = pd.read_excel(path_raw_data + file_name, index_col=0)
        temp_data = raw_data[(raw_data['분석제외 여부'] != '예외') &  # 분석제외 여부 '예외' 제외
                             (raw_data['분석제외 여부'] != '중복, 예외') &
                             (raw_data['통합 분류1'].isin(['경제>부동산', ' 경제>부동산']) |  # 통합 분류1과 통합 분류2에서 '경제>부동산'만
                              raw_data['통합 분류2'].isin(['경제>부동산', ' 경제>부동산']))]
        data = pd.concat([data, temp_data])
    # 불필요한 칼럼제거
    data.drop(['기고자', '통합 분류3', '분석제외 여부', '인물', '본문', '기관','통합 분류1', '통합 분류2',
               '사건/사고 분류1', '사건/사고 분류2', '사건/사고 분류3'], axis=1, inplace=True)
    # 칼럼 변경
    data['일자'] = pd.to_datetime(data['일자'],format='%Y%m%d')
    data['키워드'] = data['키워드'].apply(lambda cnt: cnt.split(','))
    data['특성추출(가중치순 상위 50개)'] = data['특성추출(가중치순 상위 50개)'].apply(lambda cnt: cnt.split(','))
    data['위치'] = data['위치'].where(pd.notnull(data['위치']), 'None') #자료가 없는 경우 문자열 'None' 입력
    data['위치'] = data['위치'].apply(lambda cnt: cnt.split(','))
    return data
```

```python
data = preprocess_news_data(path_raw_data)
```
